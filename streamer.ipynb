```python
from twython import Twython
from twython import TwythonStreamer
import twitter
```

```python
import credential
APP_KEY = credential.login['APP_KEY']
APP_SECRET = credential.login['APP_SECRET']
OAUTH_TOKEN = credential.login['OAUTH_TOKEN']
OAUTH_TOKEN_SECRET = credential.login['OAUTH_TOKEN_SECRET']
```

```python
twitter = Twython(APP_KEY, APP_SECRET,
                  OAUTH_TOKEN, OAUTH_TOKEN_SECRET)
```

```python
# The keyword query

QUERY = 'pokemon go'

# The file to write output as newline-delimited JSON documents
OUT_FILE = QUERY + ".json"
```

```python
tweets = []
NUM_TWEET = 1000
class MyStreamer(TwythonStreamer):
    
    def on_success(self,data):
        if data['lang'] == 'en':
            tweets.append(data)
            print "received tweet #", len(tweets)
        
        if (len(tweets) > NUM_TWEET):
            self.disconnect()
            
    def on_error(self, status_code,data):
        print status_code, data
        self.disconnect()

stream = MyStreamer(APP_KEY, APP_SECRET, OAUTH_TOKEN, OAUTH_TOKEN_SECRET)

    
```

```python
stream.statuses.filter(track=QUERY)
```

```python
import sys
import csv
from twitter import *
#-----------------------------------------------------------------------
# open a file to write (mode "w"), and create a CSV writer object
#-----------------------------------------------------------------------
csvfile = file("streamer_output.csv", "w")
csvwriter = csv.writer(csvfile)
#-----------------------------------------------------------------------
# add headings to our CSV file
#-----------------------------------------------------------------------
row = [ "date", "text", "place" ]
csvwriter.writerow(row)
```

```python
for tweet in tweets:
    date = tweet["created_at"]
    text = tweet["text"]
    text = text.encode('ascii', 'replace')
    if tweet["place"]:
        place = tweet["place"]["full_name"]
        place = place.encode('ascii', 'replace')
    else: place = "NA"
    
    # now write this row to our CSV file
    row = [ date, text, place]
    csvwriter.writerow(row)
```
